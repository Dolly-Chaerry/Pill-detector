```python
import cv2
import numpy as numpy
import matplotlib.pyplot as plt
import joblib
from feature_extractor import FeatureExtractor
from image_segmentation import ImageSegmenter
```

```python
img = cv2.imread("multiple_pill_test2.jpg")
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
segmenter = ImageSegmenter()
extracted_pills, _,_ = segmenter.segment(img)
```

```python
for pill in extracted_pills:
    plt.figure(figsize=(5,5))
    plt.imshow(pill)
    plt.show()
```

```python
svm_model = joblib.load("svm_model.joblib")
```

```python
extractor = FeatureExtractor()
for pill in extracted_pills:
    features = extractor.get_features(pill)
    predict = svm_model.predict([features])
    if predict == 0:
        print("capsule")
    else:
        print("tablet")
    
```
